```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import os, json
import torch
```

```python
model_results_path = ["../results/flan-llama-7b-ICL-10/", "../results/llama-7b-ICL-10/"]

all_tasks = [os.listdir(path) for path in model_results_path]
```

```python
df = pd.DataFrame(columns=["model", "task", "exact_match_score", "at"])

for task in all_tasks[0]:
    for model in model_results_path:
        result = json.load(open(os.path.join(model, task, "exact_match_score_results.json")))
        for at, score in result["clean_topk"]:
            df = df.append({"model": model, "task": task, "exact_match_score": score, "at": at}, ignore_index=True)
```

```python
df
```

```python
similarity = dict()

for task in df["task"].unique():    
    sub_df = df[df["task"] == task]
    
    task_scores = {}
        
    for at in sub_df["at"].unique():
        task_scores[at] = sub_df[sub_df["at"] == at]["exact_match_score"].values
    
    similarity[task] = abs(task_scores[1][0] - task_scores[1][1])
    

sorted_similarity = sorted(similarity.items(), key=lambda x: x[1], reverse=False)
sorted_similarity
    
```
